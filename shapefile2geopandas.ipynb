```
import geopandas as gpd
PATH_TO_SHAPE="/content/drive/My Drive/geo_data/h27ka12.shp"
df_shp = gpd.read_file(PATH_TO_SHAPE,encoding='SHIFT-JIS')
```

```
df_shp.head()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_axis_off()
df_chiba = df_shp[df_shp.HCODE != 8154]
df_chiba.plot(ax=ax, column='SETAI', scheme='Fisher_Jenks', k=3, cmap='OrRd', legend=True)
```
